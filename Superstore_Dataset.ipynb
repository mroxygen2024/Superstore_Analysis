## Task 1: Load & Explore Data

```python
import pandas as pd

# Load the dataset into a Pandas DataFrame, specifying 'latin1' encoding to handle potential character issues
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')

print("Dataset loaded successfully.")
```

### Display the first few rows

```python
display(df.head())
```

### Check data types and dataset structure

```python
print(df.info())
```

### Generate summary statistics

```python
display(df.describe(include='all'))
```

## Task 2: Data Cleaning

### Identify missing values

```python
# Display the number of missing values per column
print("Missing values before cleaning:")
display(df.isnull().sum())
```

### Handle missing data

```python
# Drop rows with any missing values
df.dropna(inplace=True)

print(f"Dataset shape after dropping missing values: {df.shape}")
print("Missing values after cleaning:")
display(df.isnull().sum())
```

### Convert relevant columns to correct data types

```python
# Convert 'Order Date' and 'Ship Date' to datetime objects
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

print("Data types after conversion:")
print(df.info())
```

## Task 3: Feature Engineering

### Extract Month and Year from the order date

```python
# Extract month and year from 'Order Date'
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year

print("Month and Year extracted successfully:")
display(df[['Order Date', 'Order Month', 'Order Year']].head())
```

### Create a new column based on existing numerical columns

```python
# Create a new column: 'Sales After Discount'
df['Sales After Discount'] = df['Sales'] * (1 - df['Discount'])

print("New column 'Sales After Discount' created:")
display(df[['Sales', 'Discount', 'Sales After Discount']].head())
```

### Verify whether your new column is logically correct

```python
# Verify the new column by checking its data type and descriptive statistics
print("Data type of 'Sales After Discount':", df['Sales After Discount'].dtype)
print("Descriptive statistics for 'Sales After Discount':")
display(df['Sales After Discount'].describe())
```

## Task 4: GroupBy & Aggregation

### Calculate total sales by category

```python
# Calculate total sales by 'Category'
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
sales_by_category = sales_by_category.sort_values(by='Sales', ascending=False)

print("Total Sales by Category:")
display(sales_by_category)
```

### Calculate total profit by region

```python
# Calculate total profit by 'Region'
profit_by_region = df.groupby('Region')['Profit'].sum().reset_index()
profit_by_region = profit_by_region.sort_values(by='Profit', ascending=False)

print("Total Profit by Region:")
display(profit_by_region)
```

### Identify top 5 customers based on sales

```python
# Identify top 5 customers based on 'Sales'
top_customers_by_sales = df.groupby('Customer Name')['Sales'].sum().nlargest(5).reset_index()

print("Top 5 Customers by Sales:")
display(top_customers_by_sales)
```

### Analyze sales trends over time (monthly)

```python
# Analyze monthly sales trends
monthly_sales = df.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['Order Year'].astype(str) + '-' + monthly_sales['Order Month'].astype(str).str.zfill(2)
monthly_sales = monthly_sales.sort_values(by='YearMonth')

print("Monthly Sales Trends:")
display(monthly_sales.head())
```

## Task 5: Pivot Tables

### Create pivot tables to show sales by category and region

```python
# Pivot table for sales by category and region
sales_by_category_region = pd.pivot_table(df, values='Sales', index='Category', columns='Region', aggfunc='sum')

print("Sales by Category and Region:")
display(sales_by_category_region)
```

### Create pivot tables to show sales trends by segment over time

```python
# Pivot table for sales trends by segment over time
sales_by_segment_time = pd.pivot_table(df, values='Sales', index='Order Year', columns='Segment', aggfunc='sum')

print("Sales Trends by Segment Over Time:")
display(sales_by_segment_time)
```

### Create pivot tables to show profit by sub-category

```python
# Pivot table for profit by sub-category
profit_by_subcategory = pd.pivot_table(df, values='Profit', index='Sub-Category', aggfunc='sum')
profit_by_subcategory = profit_by_subcategory.sort_values(by='Profit', ascending=False)

print("Profit by Sub-Category:")
display(profit_by_subcategory)
```

## Task 6: Data Splitting & Merging

### Split the dataset into at least two meaningful DataFrames

```python
# Split the DataFrame into two based on 'Category'
df_technology = df[df['Category'] == 'Technology'].copy()
df_office_supplies = df[df['Category'] == 'Office Supplies'].copy()

print("DataFrame for Technology products shape:", df_technology.shape)
print("DataFrame for Office Supplies products shape:", df_office_supplies.shape)

display(df_technology.head())
display(df_office_supplies.head())
```

### Merge them back into a single dataset

```python
# Merge the two DataFrames back together using pd.concat
df_merged = pd.concat([df_technology, df_office_supplies])

print("Merged DataFrame shape:", df_merged.shape)
display(df_merged.head())
```

### Ensure the merged result is correct

```python
# Verify the merged DataFrame by checking its shape and comparing unique values
print("Original DataFrame shape:", df.shape)
print("Merged DataFrame shape:", df_merged.shape)

# Check if all rows from technology and office supplies are present in the merged dataframe
original_categories = df['Category'].unique()
merged_categories = df_merged['Category'].unique()

print(f"Original Categories: {original_categories}")
print(f"Merged Categories from split: {merged_categories}")

# A more robust check: compare value counts of a relevant column, e.g., 'Row ID'
# For simplicity, we are only checking the concatenated portion, not the full original df
# If we want to check against the original df, we should merge all splits

# Verify that the number of rows for 'Technology' and 'Office Supplies' in the merged df
# matches the original counts for these categories.
original_tech_count = df[df['Category'] == 'Technology'].shape[0]
merged_tech_count = df_merged[df_merged['Category'] == 'Technology'].shape[0]

original_office_count = df[df['Category'] == 'Office Supplies'].shape[0]
merged_office_count = df_merged[df_merged['Category'] == 'Office Supplies'].shape[0]

print(f"Original Technology Count: {original_tech_count}, Merged Technology Count: {merged_tech_count}")
print(f"Original Office Supplies Count: {original_office_count}, Merged Office Supplies Count: {merged_office_count}")

if original_tech_count == merged_tech_count and original_office_count == merged_office_count:
    print("Verification successful: Row counts for split categories match after merge.")
else:
    print("Verification failed: Row counts do not match.")
```

## Task 8: From your analysis, determine:

### The most profitable category

```python
# Calculate total profit by category
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()
most_profitable_category = profit_by_category.sort_values(by='Profit', ascending=False).iloc[0]

print("Most Profitable Category:")
display(most_profitable_category)
```

### The least performing region

```python
# The 'profit_by_region' DataFrame was already created in Task 4.
# We can directly use it to find the least performing region.
least_performing_region = profit_by_region.sort_values(by='Profit', ascending=True).iloc[0]

print("Least Performing Region (by Profit):")
display(least_performing_region)
```

### Any patterns between sales and profit

```python
# To analyze patterns between sales and profit, we can look at the overall correlation
# or group by a dimension like Sub-Category and compare their sales and profit.

sales_profit_by_subcategory = df.groupby('Sub-Category')[['Sales', 'Profit']].sum().reset_index()
sales_profit_by_subcategory['Profit Ratio'] = sales_profit_by_subcategory['Profit'] / sales_profit_by_subcategory['Sales']

print("Sales and Profit by Sub-Category (sorted by Profit Ratio):")
display(sales_profit_by_subcategory.sort_values(by='Profit Ratio', ascending=False).head())
display(sales_profit_by_subcategory.sort_values(by='Profit Ratio', ascending=True).head())

# Visualize the relationship with a scatter plot
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Sales', y='Profit', data=sales_profit_by_subcategory, hue='Sub-Category', size='Sales', sizes=(20, 1000), legend='brief')
plt.title('Sales vs. Profit by Sub-Category')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')
plt.xscale('log') # Use log scale for sales to better visualize distribution
plt.yscale('symlog', linthresh=100) # Use symlog for profit to handle negative values
plt.axhline(0, color='red', linestyle='--') # Add a line at 0 profit
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.tight_layout()
plt.show()
```

### Important trends over time

```python
# The monthly sales trend was already analyzed and visualized in Task 7.
# We can also look at monthly profit trends.

monthly_profit = df.groupby(['Order Year', 'Order Month'])['Profit'].sum().reset_index()
monthly_profit['YearMonth'] = monthly_profit['Order Year'].astype(str) + '-' + monthly_profit['Order Month'].astype(str).str.zfill(2)
monthly_profit = monthly_profit.sort_values(by='YearMonth')

print("Monthly Profit Trends:")
display(monthly_profit.head())

plt.figure(figsize=(12, 6))
sns.lineplot(x='YearMonth', y='Profit', data=monthly_profit, marker='o', color='green')
plt.title('Monthly Profit Trend Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
